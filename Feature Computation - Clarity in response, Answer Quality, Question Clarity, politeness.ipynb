## Typos - Grammar check

```python
import grammar_check
tool = grammar_check.LanguageTool('en-GB')
def typo_grammar(text):
    try:
        
        if text == None:
            return -1 
        text = str(text.encode("utf8"))
        matches = tool.check(text)
        return len(matches)
    except UnicodeEncodeError:
        return 0
    except AttributeError:
        return 0
    except:
        return 0
```

```python
trial['typo_score'] = trial['Body'].apply(typo_grammar)
```

```python
from nltk import word_tokenize  
import enchant  
import re 
from enchant.checker import SpellChecker
from enchant.tokenize import EmailFilter, URLFilter
chrk= SpellChecker("en_US", filters= [EmailFilter, URLFilter])
d= enchant.Dict('en_US')
def check_spelling_score(text):
    if text == None:
        return -1
    try:
        text=str(text.encode("utf-8"))
        chrk.set_text(text)
        count=0
        for err in chrk:
            #print err.word
            count=count+1
    
        return count  
    except UnicodeEncodeError:
        return 0
    except:
        return 0
    
#outputs 1 if there is a spelling mistake
def check_spelling_binary(text):
    if text == None:
        return -1
    try:
        
        text=str(text.encode("utf-8"))
        ans = d.check(text)
        if ans==False:
            return 1
        return 0
    except UnicodeEncodeError:
        return 0
    except:
        return 0
```
