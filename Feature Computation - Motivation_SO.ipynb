```python
merged_users_posts = pd.read_csv("merged_users_posts_whole_1.csv")
merged_users_posts
```

```python
m_users_posts=merged_users_posts[['id','OwnerUserId','PostTypeId','CreationDate','Id']]
m_users_posts
```

```python
m_users_posts.groupby('id')['PostTypeId'].describe()
```

```python
p_users_posts=m_users_posts.groupby('id').size().reset_index(name='noPosts')
p_users_posts
```

```python
a_users_posts=m_users_posts[(m_users_posts['PostTypeId']==2)].groupby('id').size().reset_index(name='noAns')
a_users_posts
```

```python
q_users_posts=m_users_posts[(m_users_posts['PostTypeId']==1)].groupby('id').size().reset_index(name='noQues')
q_users_posts
```

```python
pq = p_users_posts.merge(q_users_posts, on='id', how='outer')
pq
```

```python
qap = pq.merge(a_users_posts, on='id', how='outer')
qap
```

```python
qap=qap.fillna(0)
```

```python
qap
```

```python
qap['sum']=qap['noQues']+qap['noAns']
```

```python
qap
```

```python
qap.loc[~(qap['noPosts'] == qap['sum'])]
```

```python
qap['z-score_denominator']=qap[['noAns','noQues']].sum(axis=1).pow(1./2)
```

```python
qap['z-score_numerator']=qap['noAns']-qap['noQues']
```

```python
qap
```

```python
qap['z-score']= (qap['z-score_numerator'])/(qap['z-score_denominator'])
```

```python
qap
```

```python
qap.to_csv('qap_z-score.csv')
```

```python
NUMpostsNOTequalTOsum = qap.loc[~(qap['noPosts'] == qap['sum'])]
```
