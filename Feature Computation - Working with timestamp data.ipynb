## Frequency of contributions

```python
import os
import pandas as pd
merged_users_posts = pd.read_csv("merged_users_posts.csv")
merged_users_posts
```

```python
m_users_posts=merged_users_posts[['id','OwnerUserId','PostTypeId','CreationDate','Id']]
m_users_posts
```

```python
mydate = dateutil.parser.parse('2007-04-03T21:07:12.510')
mydate
```

```python
m_users_posts
```

```python
m_users_posts['timestamp'] = m_users_posts['CreationDate'].apply(dateutil.parser.parse)
```

```python
m_users_posts
```

```python
s_users_posts = m_users_posts.sort_values(by=['OwnerUserId','timestamp'], ascending=[True,True])
s_users_posts
```

```python
#t_users_posts = s_users_posts.sort_values(by=['name','timestamp'])
s_users_posts['time_diff'] = s_users_posts.groupby('OwnerUserId')['timestamp'].diff()
```

```python
s_users_posts
```

```python
s_users_posts['time_diff_sec'] = s_users_posts.time_diff.dt.total_seconds()
```

```python
s_users_posts
```

```python
s_users_posts.to_csv('timestamped_users_posts.csv')
```
