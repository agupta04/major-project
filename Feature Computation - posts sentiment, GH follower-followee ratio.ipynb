## Spell Check 

```python
from nltk import word_tokenize  
import enchant  
import re 
from enchant.checker import SpellChecker
from enchant.tokenize import EmailFilter, URLFilter
chrk= SpellChecker("en_US", filters= [EmailFilter, URLFilter])
d= enchant.Dict('en_US')
def check_spelling_score(text):
    if text == None:
        return -1
    try:
        text=str(text.encode("utf8"))
        chrk.set_text(text)
        count=0
        for err in chrk:
            #print err.word
            count=count+1
    
        return count  
    except UnicodeEncodeError:
        return 0
    except:
        return 0
    
#outputs 1 if there is a spelling mistake
def check_spelling_binary(text):
    if text == None:
        return -1
    try:
        
        text=str(text.encode("utf8"))
        ans = d.check(text)
        if ans==False:
            return 1
        return 0
    except UnicodeEncodeError:
        return 0
    except:
        return 0
```

```python
posts['spell_score'] =posts.apply(lambda x: check_spelling_binary(str(x['Body'])),axis=1)
posts
```

## Sentiment of all posts

```python
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer

def check_pos_sentiment(text):
    try:
        text=str(text.encode("utf8"))
        blob = TextBlob(text, analyzer=NaiveBayesAnalyzer())
        return blob.sentiment[1]
    except:
        return 0
  
```

```python
trial['pos_sentiment'] =trial.apply(lambda x: check_pos_sentiment(str(x['Body'])),axis=1)
trial
```

```python
posts['pos_sentiment'] =posts.apply(lambda x: check_pos_sentiment(str(x['Body'])),axis=1)
posts
```

## Github follower / followees ratio

```python
followers = pd.read_csv("/Volumes/Seagate Backup Plus Drive/BTP_DATA/gh_use/followers.csv", header = None, encoding = "ISO-8859-1")
followers
```

```python
followers.columns = [ "user_id", "follower_id", "created_at"]  
```

```python
followers
```

```python
user_id_followers = followers.groupby(["user_id"]).size().reset_index(name='followers')
user_id_followers
```

```python
follower_id_followees = followers.groupby(["follower_id"]).size().reset_index(name='followees')
follower_id_followees
```

```python
ratio = pd.merge(left=user_id_followers,right=follower_id_followees,how='right', left_on='user_id', right_on='follower_id')
ratio.shape
```

```python
ratio
```

```python
ratio = pd.merge(left=linked,right=ratio,how='inner', left_on='gh_id', right_on='user_id')
ratio.shape
```

```python
ratio
```

```python
ratio["followers"].fillna(0, inplace= True)
```

```python
ratio
```

```python
ratio["f_ratio"] = ratio["followers"]/ratio["followees"]
```
