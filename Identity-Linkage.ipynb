```python
def find_gh_email(df):
    user_gh_dict={}
    for index, row in df.iterrows():
        user_gh_dict[row['id']]=str(row['email'])
    return user_gh_dict
```

```python
import hashlib
user_gh_list_hash=[]
for index, row in temp_df.iterrows():
    user_gh_list_hash.append(str(hashlib.md5(str(row['email'])).hexdigest()))
    #row['email_hash']=str(hashlib.md5(str(row['email'])).hexdigest())
    #user_gh_dict_hash[row['id']]=str(row['email_hash'])
temp_df['emailHash'] = user_gh_list_hash
temp_df
```
