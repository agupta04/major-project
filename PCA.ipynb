# kmeans after pca

```python
import numpy as np 
import pandas as pd 
from sklearn.decomposition import PCA 
from sklearn.cluster import KMeans 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline 
```

```python
f, ax = plt.subplots(figsize=(25, 25))
plt.title('Pearson Correlation of User Features')
# Draw the heatmap using seaborn
sns.heatmap(df_standard1.astype(float).corr(),linewidths=0.25,vmax=1.0, square=True, cmap="YlGnBu", linecolor='black', annot=True)
```

```python
# Calculating Eigenvectors and eigenvalues of Cov matirx
mean_vec = np.mean(df_standard1, axis=0)
cov_mat = np.cov(df_standard1.T)
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
```

```python
# Create a list of (eigenvalue, eigenvector) tuples
eig_pairs = [ (np.abs(eig_vals[i]),eig_vecs[:,i]) for i in range(len(eig_vals))]

# Sort from high to low
eig_pairs.sort(key = lambda x: x[0], reverse= True)

# Calculation of Explained Variance from the eigenvalues
tot = sum(eig_vals)
var_exp = [(i/tot)*100 for i in sorted(eig_vals, reverse=True)] # Individual explained variance
cum_var_exp = np.cumsum(var_exp) # Cumulative explained variance
```

```python
cum_var_exp
```

```python
## plot pca graph - 27 features
len(cum_var_exp)
```

## To ensure that useful information is not excluded, one should go for 95% or greater variance level which corresponds to about 23 components in this case

```python
plt.figure(figsize=(10, 5))
plt.bar(range(32), var_exp, alpha=0.3333, align='center', label='individual explained variance', color = 'g')
plt.step(range(32), cum_var_exp, where='mid',label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')
plt.legend(loc='best')
plt.show()
```

```python
pca = PCA(n_components=23)
x = pca.fit_transform(df_standard1)
```

```python
plt.figure(figsize = (20,20))
plt.scatter(x[:,0],x[:,1],alpha=0.8)
#plt.ylim(-10,30)
plt.show()
```

```python
# Set a 2 KMeans clustering
kmeans = KMeans(n_clusters=2)
# Compute cluster centers and predict cluster 
X_clustered = kmeans.fit_predict(x)

LABEL_COLOR_MAP = {0 : 'r',1 : 'g'}
label_color = [LABEL_COLOR_MAP[l] for l in X_clustered]

# Plot the scatter digram
plt.figure(figsize = (7,7))
plt.scatter(x[:,0],x[:,2], c= label_color, alpha=0.5) 
plt.show()
```
